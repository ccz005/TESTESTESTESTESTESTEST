```python
import pandas as pd
import numpy as np
import matplotlib
import datetime
```

```python
##Parameters
time = 60 * 3
year_from = 2017
year_to = 2017
```

```python
gold = pd.read_csv("GC #F_1.csv")
gold.head()
```

```python
silver = pd.read_csv("SI #F_1.csv")
silver.head()
```

```python
gold['DT'] = pd.to_datetime(gold['Date'] + ' ' + gold['Time'])
cols2 = ['DT', 'Open', 'High', 'Low', 'Close', 'Volume']
gold = gold[cols2]
gold.columns = ['DT', 'Open.G', 'High.G', 'Low.G', 'Close.G', 'Volume.G']
gold.head()
```

```python
silver['DT'] = pd.to_datetime(silver['Date'] + ' ' + silver['Time'])
cols2 = ['DT', 'Open', 'High', 'Low', 'Close', 'Volume']
silver = silver[cols2]
silver.columns = ['DT', 'Open.S', 'High.S', 'Low.S', 'Close.S', 'Volume.S']
silver.head()
```

```python
dataP = pd.merge(gold, silver, on = "DT", how = 'inner')
dataP.head()
```

```python
dataP['ratio'] = round(dataP['Close.G'] / dataP['Close.S'], 3)
SMA = []
sd = []
z = []
for i in range(0,len(dataP)):
    if i <= time:
        SMA_now = 0
        sd_now = 0
        z_now = 0
    else:
        SMA_now = numpy.mean(dataP[(i-time):i]['ratio'])
        sd_now = numpy.std(dataP[(i-time):i]['ratio'])
        z_now = (dataP.iloc[i]['ratio'] - SMA_now) / sd_now
        
    SMA.append(round(SMA_now, 5))
    sd.append(round(sd_now, 5))
    z.append(round(z_now, 4))

print(z[(len(z)-10):len(z)])
```

```python
SMA_array = np.asarray(SMA)
sd_array = np.asarray(sd)
z_array = np.asarray(z)

dataP['SMA'] = SMA_array
dataP['sd'] = sd_array
dataP['z'] = z_array

dataP.head()
```

```python
##Backtesting
t = time + 1
c = 1
pos = 0
prime = 0
entryt = 0
entryz = 0
history = []
temp_history = []
# weekday Monday starts as 1
```

```python
while t < 20000:
    OpenPnL = 0
    if (
    pos == 0 and
    dataP.iloc[t]['DT'].strftime("%H") <= 16 if dataP.iloc[t]['DT'].weekday() == 5 else t > 0
    ):
        if (dataP.iloc[t]['z'] < 2 and
            dataP.iloc[t]['z'] > 1.5 and
            dataP.iloc[(t-1)]['z'] > 2
        ):
            prime = -1
        elif (dataP.iloc[t]['z'] > -2 and
              dataP.iloc[t]['z'] < -1.5 and
              dataP.iloc[(t-1)]['z'] < -2
        ):
            prime = 1
    else: OpenPnL = 0
    
    if prime != 0:
        entrytime = dataP.iloc[t]['DT']
        exittime = 0
        vol_G = prime * 100 * 2
        entryprice_G = dataP.iloc[t]['Close.G']
        exitprice_G = 0
        vol_S = -prime * 5000 * 3
        entryprice_S = dataP.iloc[t]['Close.S']
        exitprice_S = 0
        profit = 0
        entryt = t
        
        pos = prime
        prime = 0
    elif (
        pos != 0 and
        t >= entryt + 1 and
        (
        OpenPnL < -3000 or t >= entryt + 30 or
        dataP.iloc[t]['z'] < 1 if pos == -1 else dataP.iloc[t]['z'] > -1
        )
    ):
        exittime = dataP.iloc[t]['DT']
        exitprice_G = dataP.iloc[t]['Close.G']
        exitprice_S = dataP.iloc[t]['Close.S']
        profit = vol_G * (exitprice_G - entryprice_G) + vol_S * (exitprice_S - entryprice_S)
        
        temp_history = [entrytime, exittime, vol_G, entryprice_G, exitprice_G, vol_S, entryprice_S, exitprice_S, profit]
        history.append(temp_history)
        
        c = c+1
        pos = 0
    
    t = t + 1
    
```

```python
history_df = pd.DataFrame(history, columns = ['entry_time', 'exit_time', 'volume_gold', 'entry_gold', 'exit_gold',
                                             'volume_silver', 'entry_silver', 'exit_silver', 'Profit'])
gross = history_df['Profit'].sum()
print(gross)
```
